```python
# Find the 95% VaR for 0.5% monthly expected return and 4% monthly volatility

import scipy.stats as stats

expected_return = 0.005
volatility = 0.04
confidence_level = 0.95
VaR_95 = stats.norm.ppf(1 - confidence_level, expected_return, volatility)
VaR_95
```

```python
# Modified critical value for 95% probability using the Cornish-Fisher expansion
import scipy.stats as stats

alpha = 0.05
z_95 = stats.norm.ppf(alpha)
skewness = -1.2  # assuming normal distribution
kurtosis = 6  # assuming normal distribution
z_95_cf = (
    z_95
    + (1 / 6) * (z_95**2 - 1) * skewness
    + (1 / 24) * (z_95**3 - 3 * z_95) * (kurtosis - 3)
    - (1 / 36) * (2 * z_95**3 - 5 * z_95) * skewness**2
)

print(f"Standard Gaussian Z: {z_95:.4f}")
print(f"Modified Cornish-Fisher Z_cf: {z_95_cf:.4f}")
```
