```python
"""
Mechanistic mapping of Calculus to Linear Algebra.
Demonstrates O(1) memory structural transformations using NumPy.

Python 3.11+ Standards:
- Uses native list/tuple typing
- Leverages vectorized numpy operations to bypass native Python loop bottlenecks
"""

import numpy as np


def build_derivative_operator(degree: int) -> np.ndarray:
    """
    Constructs a matrix operator that acts as the First Derivative
    for the basis {1, x, x^2, ..., x^n}.

    Space Complexity: O(n^2) auxiliary memory for the operator matrix.
    """
    # Initialize a square matrix of zeros
    D = np.zeros((degree + 1, degree + 1))

    # Fill the super-diagonal to map x^n -> n * x^(n-1)
    for i in range(1, degree + 1):
        D[i - 1, i] = i

    return D


def main():
    # Target Polynomial: P(x) = 5 + 2x + 3x^2 + 4x^3
    # Mapped to vector basis {1, x, x^2, x^3}
    P_vector = np.array([5, 2, 3, 4], dtype=float)
    degree = len(P_vector) - 1

    # 1. Instantiate the operator
    D_matrix = build_derivative_operator(degree)

    print("--- The Derivative Operator Matrix (D) ---")
    print(D_matrix, "\n")

    # 2. Perform the continuous calculus derivative via discrete vector dot product
    # Time Complexity: O(n^2) for standard matrix-vector multiplication
    dP_vector = D_matrix @ P_vector

    print(f"Original Vector  : {P_vector} -> 5 + 2x + 3x^2 + 4x^3")
    print(f"Derivative Vector: {dP_vector} -> 2 + 6x + 12x^2")

    # 3. Axiomatic Stress Test against native NumPy polyder
    # Note: np.poly1d processes coefficients in reverse order {x^n, ..., x, 1}
    np_poly = np.poly1d(P_vector[::-1])
    np_deriv = np.polyder(np_poly)

    # Reverse back to match our basis ordering
    assert np.allclose(dP_vector[: len(np_deriv.c)], np_deriv.c[::-1]), (
        "Matrix derivative failed!"
    )
    print(
        "\nAxiomatic Test Passed: Linear Transformation perfectly mirrors analytic calculus."
    )


if __name__ == "__main__":
    main()
```
