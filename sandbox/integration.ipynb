```python
### Symbolic Integration
```

```python
import sympy as sp
```

```python
# 1. Define the symbolic variable
x = sp.Symbol("x")
```

```python
# Case 1: x^2 from 0 to 1
expr1 = x**2
ans1 = sp.integrate(expr1, (x, 0, 1))
print(f"Integral of x^2 from 0 to 1: {ans1}")  # Output: 1/3
```

```python
# Case 2: sin(x) from 0 to 2*pi
expr2 = sp.sin(x)
ans2 = sp.integrate(expr2, (x, 0, 2 * sp.pi))
print(f"Integral of sin(x) from 0 to 2*pi: {ans2}")  # Output: 0
```

```python
# Case 3: Area of a Unit Circle using Radians
# A unit circle's upper half is y = sqrt(1 - x^2) from -1 to 1.
# Total Area = 2 * integral_{-1}^{1} sqrt(1 - x^2) dx
expr3 = sp.sqrt(1 - x**2)
ans3 = 2 * sp.integrate(expr3, (x, -1, 1))
print(f"Area of Unit Circle: {ans3}")  # Output: pi
```

```python
### Numerical Integration
```

```python
import numpy as np
from scipy import integrate
```

```python
# Case 1: x^2 from 0 to 1
f1 = lambda x: x**2
result1 = integrate.quad(f1, 0, 1)
print(
    f"Numerical integral of x^2 from 0 to 1: {result1[0]}"
)  # Output: 0.33333333333333337
```

```python
# Case 2: sin(x) from 0 to 2*pi
f2 = lambda x: np.sin(x)
result2 = integrate.quad(f2, 0, 2 * np.pi)
print(f"Numerical integral of sin(x) from 0 to 2*pi: {result2[0]}")  # Output: 0.0
```

```python
# Case 3: Area of a Unit Circle using Radians
# A unit circle's upper half is y = sqrt(1 - x^2) from -1 to 1.
# Total Area = 2 * integral_{-1}^{1} sqrt(1 - x^2) dx
f3 = lambda x: np.sqrt(1 - x**2)
result3 = integrate.quad(f3, -1, 1)
print(
    f"Numerical integral of sqrt(1 - x^2) from -1 to 1: {result3[0]}"
)  # Output: 1.5707963267948966
print(f"Area of Unit Circle: {2 * result3[0]}")  # Output: 3.141592653589793
```
