```python
import numpy as np

np.set_printoptions(precision=4, suppress=True)

%matplotlib inline

seed = 42
```

```python
p = 0.3
sample_size = [10, 100, 1_000, 10_000, 100_000, 1_000_000]
rng = np.random.default_rng(seed)

for n in sample_size:
    X = rng.binomial(n, p=p)

    expected_value = n * p
    percentage_difference = (X - expected_value) / expected_value * 100
    print(
        f"Sample size: {n}, X = {X}, Expected Value = {expected_value}, Percentage difference from expected value: {percentage_difference}%"
    )
```
