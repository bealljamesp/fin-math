```python
# 1-23

import numpy as np

seed = 42
rng = np.random.default_rng(seed=seed)
```

```python
n_rolls = 10_000

roll = rng.integers(low=1, high=7, size=n_rolls)

A = [2, 4, 6]
B = [1, 2, 3, 4]

is_in_A = np.isin(roll, A)
is_in_B = np.isin(roll, B)
is_in_A_and_B = np.logical_and(is_in_A, is_in_B)

p_A = np.mean(is_in_A)
p_B = np.mean(is_in_B)
p_AB = np.mean(is_in_A_and_B)

print(f"P(A) = {p_A:.4f}")
print(f"P(B) = {p_B:.4f}")
print(f"P(A and B) = {p_AB:.4f}")
print(f"P(A) * P(B) = {p_A * p_B:.4f}")
```
