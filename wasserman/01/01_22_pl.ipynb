```python
import numpy as np
```

```python
seed = 42
rng = np.random.default_rng(seed)
```

```python
probability_list = np.array([0.3, 0.5, 0.7])
sample_size = np.array([10, 100, 1_000, 10_000])
```

```python
for probability in probability_list:
    print(f"Probability: {probability}")
    for size in sample_size:
        X = rng.binomial(size, p=probability)
        print(
            f"Sample Size: {size:<9_} | "
            f"X = {X:<9_} |"
            f"np = {size * probability:<9_} |"
            f"Percent Error: {abs(X - size * probability) / (size * probability) * 100:.4f}%"
        )
    print()
```

```python
import numpy as np

probability_listv = [0.3, 0.5, 0.7]
sample_sizev = np.array([10, 100, 1_000, 10_000, 100_000, 1_000_000])
rng = np.random.default_rng(seed=422)

# Outer Loop for Logic (Iterating over macro probability regimes)
for probability in probability_listv:
    print(f"Probability Regime: {probability}")

    # MATH IS VECTORIZED: No inner loop for sample size.
    # NumPy draws all 6 binomial experiments simultaneously.
    X_vector = rng.binomial(sample_sizev, p=probability)
    print(f"X Vector: {X_vector}, np Vector: {sample_size * probability}")

    # Expected values and errors are calculated instantly as vectors
    expected_vector = sample_sizev * probability
    percent_errors = abs(X_vector - expected_vector) / expected_vector * 100

    # LOGIC LOOP: We still need a loop to display the data sequentially
    for size, X, expected, error in zip(
        sample_sizev, X_vector, expected_vector, percent_errors
    ):
        print(
            f"Sample Size: {size:<12_} | "
            f"X: {X:<10_} | "
            f"np = {float(expected):<12_.4f} | "  # Cast element to float for printing
            f"Percent Error: {error:.4f}%"
        )
    print()
```
