### Bayes' Theorem

$$P(A_i|B) = \frac{P(B|A_i)P(A_i)}{\sum_{j=1}^{n} P(B|A_j)P(A_j)}$$

_Given:_

$A_1$ = 0.7 (spam), $A_2$ = 0.2 (low priority), $A_3$ = 0.1 (high priority)

- _notice $A_1 + A_2 + A_3$ = 1_

$B$ (email contains word 'free')

Per prior experience, $P(B|A_1)$ = 0.9, $P(B|A_2)$ = 0.01, $P(B|A_3)$ = 0.01

_Find:_

If email contains 'free', what are the chances it is spam?

$P(A_1|B) = \frac{(0.9*0.7)}{((0.9*0.7)+(0.01*0.2)+(0.01*0.1)} = 0.995$
