If $\mathbb{P}(A \mid B) > \mathbb{P}(A)$, the events have positive alignment (knowing $B$ makes $A$ more likely).

If $\mathbb{P}(A \mid B) < \mathbb{P}(A)$, the events have negative alignment (knowing $B$ makes $A$ less likely).

If $\mathbb{P}(A \mid B) = \mathbb{P}(A)$, they have zero alignment—they are independent.
